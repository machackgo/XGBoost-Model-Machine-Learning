## Part 1 - Data Preprocessing

### Importing the dataset

```
import pandas as pd
dataset = pd.read_csv("churn_modelling.csv")
```

```
# suppose you want any specific number of rows in the data set the just enter any number in the head bracket
# like i have entered 10 to see 10 rows of my  dataset if you dont enter any number by default it will give youu 5 rows of your
# dataset
dataset.head(10)
```

### Checking missing data

```
dataset.info()
```

### Handling categorical variables

CustomerId and Surname columns

```
# we shouid drop this 2 column because this 2 coulmn have no coorelation with our dependent variable
# means without this 2 coulmn we can make prediction also
```

```
dataset.drop(['CustomerId', 'Surname'], axis = 1, inplace = True)
```

```
dataset.head(10)
```

Geography column

```
dataset['Geography'].unique()
```

```
geography_dummies = pd.get_dummies(dataset['Geography'], drop_first = True)
```

```
geography_dummies
```

```
dataset = pd.concat([geography_dummies, dataset], axis = 1)
```

```
dataset['Germany'] = dataset['Germany'].apply(lambda x:0 if x == False else 1)
```

```
dataset['Spain'] = dataset['Spain'].apply(lambda x:0 if x == False else 1)
```

```
dataset.head()
```

```
dataset.drop(['Geography'], axis = 1, inplace = True)
```

```
dataset.head()
```

Gender column

```
dataset['Gender'].unique()
```

```
dataset['Gender'] = dataset['Gender'].apply(lambda x:0 if x == 'Female' else 1)
```

```
dataset.head(10)
```

### Creating the Training Set and the Test Set

Getting the inputs and output

```
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

Getting the Training Set and the Test Set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Part 2 - Building and training the model

### Building the model

```
import xgboost
model = xgboost.XGBClassifier(max_depth = 4, learning_rate = 0.1, n_estimators = 100, booster = 'dart')
```

### Training the model

```
model.fit(X_train, y_train)
```

### Inference

```
y_pred = model.predict(X_test)
```

```
y_pred
```

```
y_test
```

## Part 3: Evaluating the model

### Making the Confusion Matrix

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
```

### k-Fold Cross Validation

```
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = model,
                      X = X,
                      y = y,
                      cv = 10,
                      scoring = 'accuracy'
                      )
```

```
print("Average Accuracy : {:.3f}".format(accuracies.mean()))
print("Stamdard Deviation : {:.3f}".format(accuracies.std()))
```
