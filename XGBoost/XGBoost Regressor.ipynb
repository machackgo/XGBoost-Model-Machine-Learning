## Part 1 - Data Preprocessing

### Importing the dataset

```python
import pandas as pd
dataset = pd.read_csv('insurance.csv')
```

```python
dataset.head()
```

### Checking missing data

```python
# so here by using info() we can see that  we are not havig any missing data in oor dataset if we would haver that then we will use the concept of
# taking care opf missing data which we have done in ML-A-Z course
dataset.info()
```

### Handling categorical variables

```python
# so here we are having multiple catogrial variable so thats why we are using this concept
# if we would have only one catogerial varible then we will usae that lable encoding method which we have use in
# ML-A-Z course in data preprocessing
```

Sex column

```python
# so with the help of unique key word we can determine how many catogeries are there that we need to encode
#  so as you can see here we have 2 catogeries here that we need to encode
dataset['sex'].unique()
```

```python
dataset['sex'] = dataset['sex'].apply(lambda x:0 if x == 'female' else 1)
```

```python
dataset.head()
```

Smoker column

```python
dataset['smoker'] = dataset['smoker'].apply(lambda x:0 if x == 'no' else 1)
```

```python
dataset.head()
```

Region column

```python
# so here we can see that we are having 4 catogeries to encode beacuse of that we are using dummy variable
```

```python
dataset['region'].unique()
```

```python
# if in a single coulmn if we are having  more than 2 catogerical variable then we will use this method
#  in which we are taking get_dummies method inside which we are specifying the column name
# if we want to remove the redundent column or dummy variable trap then we will use drop_first = TRUE

# we are specifying drop_first = TRUE because as you can see in our region coulmn we are having 4 catogerical vairables
# but we should encode only 3 catogerical variables and drop first one to avoide dummy variable trap
# thats why we are using drop_first = TRUE
```

```python
region_dummies = pd.get_dummies(dataset['region'], drop_first = True)
```

```python
region_dummies
```

```python
# now we are adding this dummy variable coulms into our dataset by using concact function
# so here at firsty we have specify our dummy varible name after that the dataste name and axis = 1
# axis = 1 means that we want to do cocatination in column wise and if we want to do concatination row wise then
# we can give axis = 0
```

```python
dataset = pd.concat([region_dummies, dataset], axis = 1)
```

```python
dataset.head()
```

```python
# so here we know that machine learning model will not take the data in the string form so we wil need to convert
#  that string data into  nunbers or int so that our model can accept it
```

```python
dataset['northwest'] = dataset['northwest'].apply(lambda x:0 if x == False else 1)
dataset['southeast'] = dataset['southeast'].apply(lambda x:0 if x == False else 1)
dataset['southwest'] = dataset['southwest'].apply(lambda x:0 if x == False else 1)
```

```python
dataset.head()
```

```python
#  so here with the help of region column we have taken out 3 column nothwest,southeast,southweat so we dont need that
# region coulmn any more so we will dop that coulmn
# for that we need to specify is at first coulmn name and axis = 1 that means a column and inplace = True that means
# this condition is true waht we are specifying here
```

```python
dataset.drop('region', axis = 1, inplace = True)
```

```python
dataset.head()
```

### Creating the Training Set and the Test Set

Getting the inputs and output

```python
#  here we are supperatimng the independent variabale and dependent variable
```

```python
X = dataset.iloc[:, : -1].values
```

```python
X
```

```python
y = dataset.iloc[:, -1].values
```

```python
y
```

Getting the Training Set and the Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Part 2 - Building and training the model

### Building the model

```python
#  so here we arer building the model using xgboost class hence this xgboost class was build recently
#  so thats why its does not contain he sklearn library beacuse sklearn library was build was before then xgboost
```

```python
# max_depth,learning_rate,n_estameter we are doing this because to evaluate the perforamnce the model
# so you can increase or decrteate the value of thsi as per you get the better performanve of th model
# in the last k-fold section
```

```python
import xgboost
model  = xgboost.XGBRegressor(max_depth = 2, learning_rate = 0.1, n_estameter = 100)
```

### Training the model

```python
model.fit(X_train, y_train)
```

### Inference

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

## Part 3: Evaluating the model

### R-Squared

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
```

```python
r2
```

### Adjusted R-Squared

```python
# k = number of features or number of independent variables
# n = total number of observatiuon or total numbner of customers in the test set
# shape will give the number of rows and number of coulmn
# in k 1 will give us the number of column
# andf 0 will giver the number  of rows
#  the formula for adjeent r2 iis 1-(1-r2)*(n-1)/(n-k-1)
```

```python
k = X_test.shape[1]
n = X_test.shape[0]
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
```

```python
adj_r2
```

### k-Fold Cross Validation

```python
# k-fold is the more prefound method of evaluation beacuse unlike r2 or adjemct r2 which
# which do the evaluation on one 1 train test split this k-fold will give us evaluation on 10
# train test split in which we call them train test flods it is like we have 10 differnt pairs of train test split
# and meausre of r2 then we will take mean of that r2  train test flods which givew us best performance as
# compareto r2 and adjenct r2
# for more information  about thern see 1:58:15 in XGBoost for Regression vedio (see it)
```

```python
# cross_val_score will give us the evaluation in the form of a list which having 10 elements in it
#  insted of means of that 10 values of elements
# model = XGBoost model which we have trained our data on
# X                     =   X
# name of the parameter | independent values  of our dataset
# y                     =   y
# name of the parameter | dependent values  of our dataset
# cv = number of train test flods you want to create
```

```python
from sklearn.model_selection import cross_val_score
r2s = cross_val_score(estimator = model,
                      X = X,
                      y = y,
                      cv = 10,
                      scoring = 'r2'
                      )
```

```python
# to display our means in standard devetation formate we will use {:.2f}
# because we would like 2 decemails after comma we will give 2f and f means float
#  now to convert that r2 list of 10 into mean we will give a foramt to it like format(r2.mean())
# std  = standard devetation of our list of 10 elements
```

```python
print("Average R-Square: {:.3f}".format(r2s.mean()))
print("Standard Deviation: {:.3f}".format(r2s.std()))
```
